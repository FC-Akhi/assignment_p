## **Question 1:**
Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word.

*Note*
- You have to write at least 2 additional test cases in which your program will run successfully and provide an explanation for the same.

- Example input - string = “write write write all the number from from from 1 to 100”

- Example output - 5

- Explanation - From the given string we can note that the most frequent words are “write” and “from” and the maximum value of both the values is “write” and its corresponding length is 5

```python
# calculate highest frequency word length 
def highest_frequency_word_length(string):
    """This function will help to calculate highest frequency word length
       input: string
       returns: word length which appears highest in the string"""
    
    # split the string into words
    words = string.split()
    
    # create a dictionary to store the frequency of each word
    frequency = {}
    for word in words:
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
    
    # find the highest frequency
    max_frequency = max(frequency.values())
    
    # find the length of the highest-frequency word
    for word, freq in frequency.items():
        if freq == max_frequency:
            return len(word)
        
        
```

```python
# Test case 1
string = "write write write all the number from from from 1 to 100"
print(highest_frequency_word_length(string))  # Output: 5
```

```python
# Test case 2
string = "the cat in the hat sat on the mat"
print(highest_frequency_word_length(string))  # Output: 3
```

```python
# Test case 3
string = "the quick brown fox jumps over the lazy dog"
print(highest_frequency_word_length(string))  # Output: 3
```

## **Question 5**

Write a program to download the data from the given API link and then extract the following data with
proper formatting

**Excepted Output Data Attributes -**

● Original image link - string

```python
import requests

# API endpoint URL
api_url = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"

# Send GET request to the API
response = requests.get(api_url)

# Check if the request was successful
if response.status_code == 200:
    # Extract the JSON data from the response
    data = response.json()

    # Extract the show details
    show_id = data["id"]
    show_url = data["url"]
    show_name = data["name"]

    # Print the show details
    print("Show ID:", show_id)
    print("Show URL:", show_url)
    print("Show Name:", show_name)

    # Extract and print episode details
    episodes = data["_embedded"]["episodes"]
    for episode in episodes:
        episode_id = episode["id"]
        episode_season = episode["season"]
        episode_number = episode["number"]
        episode_type = episode["type"]
        episode_airdate = episode["airdate"]
        episode_airtime = episode["airtime"]
        episode_runtime = episode["runtime"]
        episode_rating = episode["rating"]["average"]
        episode_summary = episode["summary"]
        episode_summary = episode_summary.strip("<p>").strip("</p>")
        episode_image_medium = episode["image"]["medium"]
        episode_image_original = episode["image"]["original"]

        # Print the episode details
        print("--- Episode", episode_number, "of Season", episode_season, "---")
        print("Episode ID:", episode_id)
        print("Episode Type:", episode_type)
        print("Airdate:", episode_airdate)
        print("Airtime:", episode_airtime)
        print("Runtime:", episode_runtime)
        print("Average Rating:", episode_rating)
        print("Summary:", episode_summary)
        print("Medium Image Link:", episode_image_medium)
        print("Original Image Link:", episode_image_original)

else:
    # Request failed, print the status code
    print("Error:", response.status_code)
```

## **Question 8**
Using the data from Question 5, write code the analyze the data and answer the following questions Note -
1. Draw plots to demonstrate the analysis for the following questions and better visualizations
2. Write code comments wherever required for code understanding

```python
import matplotlib.pyplot as plt

# API endpoint URL
api_url = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"

# Send GET request to the API
response = requests.get(api_url)

# Check if the request was successful
if response.status_code == 200:
    # Extract the JSON data from the response
    data = response.json()

    # Extract episode details
    episodes = data["_embedded"]["episodes"]

    # Question 1: Get all the overall ratings for each season and compare them using plots
    season_ratings = {}  # Dictionary to store season ratings
    for episode in episodes:
        season = episode["season"]
        rating = episode["rating"]["average"]
        if season not in season_ratings:
            season_ratings[season] = []
        season_ratings[season].append(rating)

    # Plotting season ratings
    seasons = list(season_ratings.keys())
    ratings = [sum(ratings_list) / len(ratings_list) for ratings_list in season_ratings.values()]

    plt.figure(figsize=(8, 5))
    plt.plot(seasons, ratings, marker='o')
    plt.xlabel('Season')
    plt.ylabel('Average Rating')
    plt.title('Average Ratings for Each Season')
    plt.grid(True)
    plt.show()

    # Question 2: Get all the episode names with average rating > 8 for each season
    high_rated_episodes = {}  # Dictionary to store high-rated episode names by season
    for episode in episodes:
        season = episode["season"]
        rating = episode["rating"]["average"]
        if rating > 8:
            if season not in high_rated_episodes:
                high_rated_episodes[season] = []
            high_rated_episodes[season].append(episode["name"])

    # Printing high-rated episode names by season
    for season, episode_names in high_rated_episodes.items():
        print("Season", season, "High-Rated Episodes:")
        for name in episode_names:
            print(name)
        print()

    # Question 3: Get episode names that aired before May 2019
    episodes_before_2019 = [episode["name"] for episode in episodes if episode["airdate"] < "2019-05"]

    # Printing episode names aired before May 2019
    print("Episodes Aired Before May 2019:")
    for name in episodes_before_2019:
        print(name)
    print()

    # Question 4: Get episode name with the highest and lowest rating for each season
    season_highest_ratings = {}  # Dictionary to store highest-rated episode names by season
    season_lowest_ratings = {}  # Dictionary to store lowest-rated episode names by season
    for episode in episodes:
        season = episode["season"]
        rating = episode["rating"]["average"]
        episode_name = episode["name"]
        if season not in season_highest_ratings:
            season_highest_ratings[season] = {"name": episode_name, "rating": rating}
            season_lowest_ratings[season] = {"name": episode_name, "rating": rating}
        else:
            if rating > season_highest_ratings[season]["rating"]:
                season_highest_ratings[season] = {"name": episode_name, "rating": rating}
            if rating < season_lowest_ratings[season]["rating"]:
                season_lowest_ratings[season] = {"name": episode_name, "rating": rating}

    # Printing episode names with the highest and lowest rating by season
    for season, episode_data in season_highest_ratings.items():
        print("Season", season, "Highest-Rated Episode:")
        print("Episode Name:", episode_data["name"])
        print("Rating:", episode_data["rating"])
        print()

    for season, episode_data in season_lowest_ratings.items():
        print("Season", season, "Lowest-Rated Episode:")
        print("Episode Name:", episode_data["name"])
        print("Rating:", episode_data["rating"])
        print()

    # Question 5: Get the summary for the most popular (highest ratings) episode in every season
    season_most_popular = {}  # Dictionary to store most popular episode summaries by season
    for episode in episodes:
        season = episode["season"]
        rating = episode["rating"]["average"]
        summary = episode["summary"]
        if season not in season_most_popular:
            season_most_popular[season] = {"summary": summary, "rating": rating}
        else:
            if rating > season_most_popular[season]["rating"]:
                season_most_popular[season] = {"summary": summary, "rating": rating}

    # Printing summaries for the most popular episode by season
    for season, episode_data in season_most_popular.items():
        print("Season", season, "Most Popular Episode Summary:")
        print(episode_data["summary"])
        print()

else:
    # Request failed, print the status code
    print("Error:", response.status_code)
```
